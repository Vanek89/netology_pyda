```python
import pandas as pd
# Задание 1
# Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
#     - оценка 2 и меньше - низкий рейтинг
#     - оценка 4 и меньше - средний рейтинг
#     - оценка 4.5 и 5 - высокий рейтинг
# Результат классификации запишите в столбец class
data = pd.read_csv('ratings.csv')
data.head()
```

```python
def classify(row):
    if row['rating'] <= 2:
        return 'низкий рейтинг'
    elif row['rating'] <= 4:
        return 'средний рейтинг'
    elif 4.5 <= row['rating'] <= 5:
        return 'высокий рейтинг'
data['class'] = data.apply(classify, axis=1)
data.head()
```

```python
# Задание 2
# Используем файл keywords.csv.
# Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность 
# определенному региону. Т. е. если поисковый запрос содержит название города региона, 
# то в столбце 'region' пишется название этого региона. Если поисковый запрос не содержит названия города, 
# то ставим 'undefined'.
# Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:
geo_data = {
     'Центр': ['москва', 'тула', 'ярославль'],
     'Северо-Запад': ['петербург', 'псков', 'мурманск'],
     'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
# Результат классификации запишите в отдельный столбец region.
geo = pd.read_csv('keywords.csv')
geo.head()
```

```python
def geo_classify(row):
    for regions, cities in geo_data.items():
        if row in cities:
            return regions
        else:
            return 'undefined'
        
geo['region'] = geo['keyword'].apply(geo_classify)
```

```python
geo['region'].value_counts()
```

```python
#  Пункт 1
years = []
x = 1950
while x != 2011:
    years.append(x)
    x += 1
print(years)
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
ratings.head()
```

```python
joined = ratings.merge(movies, on='movieId')
joined.head()
```

```python
#  Пункт 2
def production_year(row):
    for year in years:
        if str(year) in row:
            return year
    return 1900

joined['year'] = joined['title'].apply(production_year)
joined.tail()
```

```python
joined.groupby('year').mean()[['rating']].sort_values(by='rating', ascending=False).head()
```

```python
def mean_rating(data):
    return data.rating.mean()

joined.groupby('year').apply(mean_rating).sort_values(ascending=False).head()
```
