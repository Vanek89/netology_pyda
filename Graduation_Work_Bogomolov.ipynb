### 2. Рассчитать основные статистики для переменных (среднее,медиана,мода,мин/макс,сред.отклонение).


```python
def basic_metrics(df, column):
    print('Среднее для', column, '=', statistics.mean(df[column]))
    print('Медиана для', column, '=', statistics.median(df[column]))
    print('Мода для', column, '=', statistics.mode(df[column]))
    print('Минимум для', column, '=', min(df[column]))
    print('Максимум для', column, '=', max(df[column]))
    print('Стандартное отклонение для', column, '=', statistics.pstdev(df[column]))
```

Основные статистики для - satisfaction_level - Уровень удовлетворенности

```python
basic_metrics(data, 'satisfaction_level')
```

### 5.С использованием t-test статистики проверить взаимосвязи между несколькими переменными.


```python
low_salary = data[data['salary']== 'low']['satisfaction_level']
medium_salary = data[data['salary']== 'medium']['satisfaction_level']
high_salary = data[data['salary']== 'high']['satisfaction_level']

stats.f_oneway(low_salary, medium_salary, high_salary)
```

Уровень удовлетворенности отличается в зависимости от уровня заработной платы

```python
low_salary_project = data[data['salary']== 'low']['number_project']
medium_project = data[data['salary']== 'medium']['number_project']
high_project = data[data['salary']== 'high']['number_project']


stats.f_oneway(low_salary_project, medium_project, high_project)
```

По количеству проектов выборки не отличаются в зависимости от уровня заработной платы

```python
low_num = data[data['number_project'] < 4]['satisfaction_level']
high_num = data[data['number_project'] > 4]['satisfaction_level']

print (stats.ttest_ind(low_num,high_num))
```

Уровень удовлетворенности отличается в зависимости от количества проектов

```python
low_salary_exp = data[data['salary']== 'low']['time_spend_company']
medium_exp = data[data['salary']== 'medium']['time_spend_company']
high_exp = data[data['salary']== 'high']['time_spend_company']
['number_project']

stats.f_oneway(low_salary_exp, medium_exp, high_exp)
```

### 10.Проверить несколько гипотез для данных, чтобы понять почему сотрудники покидают компанию.


```python
data['left'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

y = data.left
X = data.loc[:, ['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours',\
             'time_spend_company', 'Work_accident', 'promotion_last_5years']]

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestClassifier

# создаем модель деревья решений
model=RandomForestClassifier(n_estimators=100)

# обучаем модель
model.fit(X_train,y_train)

headers = list(X_train.columns.values)

feature_imp = pd.Series(model.feature_importances_,index=headers).sort_values(ascending=False)

f, ax = plt.subplots(figsize=(10, 10))
sns.barplot(x=feature_imp, y=feature_imp.index)

plt.xlabel('Важность атрибутов')
plt.ylabel('Атрибуты')
plt.title("Наиболее важные атрибуты")
plt.legend()
plt.show()
```

```python
model.score(X_test, y_test)
```

Тремя основными факторами увольнения из компании являются - Уровень удовлетворенности, стаж в компании и количество проектов. 
Также существенно влияет количество рабочих часов в месяц.
